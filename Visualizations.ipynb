```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
# Loading the data
df1 = pd.read_csv('Query_Outputs/q1.csv')
df2 = pd.read_csv('Query_Outputs/q2.csv')
df3 = pd.read_csv('Query_Outputs/q3.csv')
df4 = pd.read_csv('Query_Outputs/q4.csv')
df5 = pd.read_csv('Query_Outputs/q5.csv')
df6 = pd.read_csv('Query_Outputs/q6.csv')
df7 = pd.read_csv('Query_Outputs/q7.csv')
df8 = pd.read_csv('Query_Outputs/q8.csv')
```

```python
# Visualization for Query 1
plt.figure(figsize=(10,6))
sns.barplot(x='percentage', y='country_name', data=df1)
plt.title('Top 5 Countries with Highest Renewable Investment in 2021')
plt.xlabel('Percentage')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='change', y='country_name', data=df3)
plt.title('Top 10 Countries with Highest Change in Renewable Investment (1990-2020)')
plt.xlabel('Change')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(15,10))
palette = sns.color_palette("Blues", len(df4))
rank = df4['average_percentage'].argsort().argsort()
sns.barplot(x='average_percentage', y='category_name', data=df4, palette=np.array(palette)[rank])
plt.title('Average Investment in Different Energy Types in 2020')
plt.xlabel('Average Percentage')
plt.ylabel('Energy Type')
for i, v in enumerate(df4['average_percentage']):
    plt.text(v + 0.1, i + .25, str(v), color='black', fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(20,10))
df5 = df5.sort_values('percentage')
palette = sns.color_palette("Greens", len(df5))
rank = df5['percentage'].argsort().argsort()
plt.barh(df5['country_name'], df5['percentage'], color=np.array(palette)[rank])
plt.xlabel('Percentage')
plt.ylabel('Country')
plt.title('Top 40 Countries with Highest Solid Biofuels Usage in 2020')
for i, v in enumerate(df5['percentage']):
    plt.text(v + 0.5, i + .25, str(v), color='blue', fontweight='bold')
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure(figsize=(15,10))
width = 0.35
labels = df8['country_name']
x = np.arange(len(labels))

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, df8['consumption_1990'], width, label='1990', color='skyblue')
rects2 = ax.bar(x + width/2, df8['consumption_2020'], width, label='2020', color='royalblue')

ax.set_xlabel('Country')
ax.set_ylabel('Consumption')
ax.set_title('Change in Total Renewable Energy Consumption from 1990 to 2020')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

for rect in rects1 + rects2:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
            '%d' % int(height),
            ha='center', va='bottom')

ax.yaxis.set_ticks(np.arange(0, max(df8['consumption_2020'])+10, 10))

plt.show()
```
